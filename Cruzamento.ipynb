```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Carregar os Dados ---
# Certifique-se de que o nome do arquivo corresponde exatamente ao que você subiu
file_path = 'Cruzamento - Página1.csv'
try:
    # Tenta ler com o separador padrão (vírgula)
    df = pd.read_csv(file_path)
except Exception as e:
    # Se der erro, tenta ler com ponto e vírgula (comum no Brasil)
    print(f"Erro ao ler com vírgula: {e}. Tentando com ponto e vírgula...")
    df = pd.read_csv(file_path, sep=';')
```

```python

# --- 2. Criar o Gráfico ---
# Define o tamanho da figura para melhor visualização
fig, ax1 = plt.subplots(figsize=(14, 8))

# --- Eixo Y da Esquerda (ax1) para o Crédito Rural ---
# Usaremos barras para o crédito, facilitando a diferenciação
color_credito = 'steelblue'
ax1.set_xlabel('Ano', fontsize=12)
ax1.set_ylabel('Crédito Rural (R$ Bilhões)', color=color_credito, fontsize=12)
ax1.bar(df['Ano'], df['Total de Crédito Rural (R$ Bilhões de 2023)'], color=color_credito, label='Crédito Rural (Bilhões R$)')
ax1.tick_params(axis='y', labelcolor=color_credito)
ax1.grid(axis='y', linestyle='--', alpha=0.7) # Adiciona uma grade suave

# --- Eixo Y da Direita (ax2) para os PIBs ---
# Cria um segundo eixo que compartilha o mesmo eixo X
ax2 = ax1.twinx()
color_pib_brasil = 'firebrick'
color_pib_agro = 'darkgreen'

ax2.set_ylabel('PIB (R$ Bilhões)', fontsize=12)
# Plota o PIB do Brasil e do Agronegócio como linhas
line1, = ax2.plot(df['Ano'], df['PIB Brasil (R$ bilhões)'], color=color_pib_brasil, marker='o', linestyle='--', label='PIB Brasil (Bilhões R$)')
line2, = ax2.plot(df['Ano'], df['PIB Agronegócio (R$ bilhões)'], color=color_pib_agro, marker='s', linestyle='-', label='PIB Agronegócio (Bilhões R$)')
ax2.tick_params(axis='y')


# --- 3. Finalizar e Mostrar o Gráfico ---
# Adiciona um título geral
plt.title('Crédito Rural vs. Evolução do PIB (Brasil e Agronegócio)', fontsize=16, pad=20)

# Combina as legendas de ambos os eixos em um único local
lines = [line1, line2]
ax1.legend(handles=lines, loc='lower right')
ax1.get_legend().remove() # Remove a legenda duplicada se houver
fig.legend(loc='lower center', bbox_to_anchor=(0.5, 0.95), ncol=3, frameon=False)


# Melhora o layout para evitar que os rótulos se sobreponham
fig.tight_layout(rect=[0, 0, 1, 0.96])

# Exibe o gráfico
plt.show()
```
